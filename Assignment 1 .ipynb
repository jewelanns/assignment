### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['SL'],y=df['PL'])
plt.title("The classification of iris types",fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sepal Length',fontsize=18)
plt.ylabel('Petal Length',fontsize=18)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['SL'],y=df['PL'],hue=df['Classification'])
plt.title("Sepal length vs Petal length",fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sepal Length',fontsize=18)
plt.ylabel('Petal Length',fontsize=18)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['SW'],y=df['PW'],hue=df['Classification'])
plt.title("Sepal width vs Petal width",fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sepal Width',fontsize=18)
plt.ylabel('Petal Width',fontsize=18)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['Classification'])
plt.title("The classification of iris types",fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Classification',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.tight_layout()
```
