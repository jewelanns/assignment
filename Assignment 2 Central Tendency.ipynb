```python
l=df.groupby('gender')['math score'].describe()
m=df.groupby('gender')['reading score'].describe()
n=df.groupby('gender')['writing score'].describe()
print(l,'\n',m,'\n',n)
```

### On the basis of gender

```python
l=df.groupby('gender')['math score'].describe()
m=df.groupby('gender')['reading score'].describe()
n=df.groupby('gender')['writing score'].describe()
print('Math score','\n',l,'\n','\n','Reading score','\n',m,'\n','\n','Writing score','\n',n)
```

```python
ld=l[['mean','std']]
md=m[['mean','std']]
nd=n[['mean','std']]
print('Math score','\n',ld,'\n','\n','Reading score','\n',md,'\n','\n','Writing score','\n',nd)
```

### On the basis of Test preparation course

```python
p=df.groupby('test preparation course')['math score'].describe()
q=df.groupby('test preparation course')['reading score'].describe()
r=df.groupby('test preparation course')['writing score'].describe()
print('Math score','\n',p,'\n','\n','Reading score','\n',q,'\n','\n','Writing score','\n',r)
```

```python
pd=p[['mean','std']]
qd=q[['mean','std']]
rd=r[['mean','std']]
print('Math score','\n',pd,'\n','\n','Reading score','\n',qd,'\n','\n','Writing score','\n',rd)
```
